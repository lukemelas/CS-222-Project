```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20,10)
```

```python
def load_image( infilename ) :
    img = Image.open( infilename )
    img.load()
    data = np.asarray( img, dtype="int32" )
    return data
```

```python
orig = load_image('../data/coco-subset-val/fake_class/COCO_train2014_000000000025.jpg')
orig_0 = load_image('../output/orig_00001.png')
comp_0 = load_image('../output/comp_00001.png')

plt.imshow(orig); plt.show()
plt.imshow(orig_0); plt.show()
plt.imshow(comp_0); plt.show()
```
